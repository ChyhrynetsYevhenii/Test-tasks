```python
import matplotlib.pyplot as plt
import numpy as np

#    
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.log(x + 1)  #  1,    

#     
plt.figure(figsize=(10, 8))

#   
plt.subplot(2, 2, 1)  # (, ,  )
plt.plot(x, y1, '-g')
plt.title('sin(x)')

#   
plt.subplot(2, 2, 2)
plt.plot(x, y2, '-b')
plt.title('cos(x)')


#   
plt.subplot(2, 2, 3)
plt.plot(x, y3, '-r')
plt.ylim(-10, 10)  #   tan(x),    
plt.title('tan(x)')

#   
plt.subplot(2, 2, 4)
plt.plot(x, y4, '-k')
plt.title('log(x + 1)')

#    ,      
plt.tight_layout()

plt.show()
```

```python
df = pd.read_csv('ds_salaries.csv')
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
 df.experience_level =  df.experience_level.replace('EN', 'Entry-level/Junior')
 df.experience_level =  df.experience_level.replace('MI', 'Mid-level/Midle')
 df.experience_level =  df.experience_level.replace('SE', 'Senior-level/Senior')
 df.experience_level =  df.experience_level.replace('EX', 'Executive-level/Team-lead')
```

```python
ex_level = df.experience_level.value_counts()
```

```python
fig = px.treemap(ex_level, path=[ex_level.index], values=ex_level.values)

fig.show()
```

```python
top15_job_titles = df.job_title.value_counts()[:15]
```

```python
fig = px.bar(y=top15_job_titles.values, x=top15_job_titles.index, text = top15_job_titles.values )

fig.show()
```

```python
emp_type = df.employment_type.value_counts()
```

```python
emp_names = ['Full-Time', 'Part-Time', 'Contract', 'Freelance']

fig = px.bar(x = emp_names, y=emp_type.values, text = emp_type.values)

fig.show()
```
